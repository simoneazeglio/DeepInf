```python
import torch 
import networkx as nx
import numpy as np 
import matplotlib.pyplot as plt
```

```python
X = torch.load("plotting.pt", map_location='cpu')
```

```python
X.shape
```

```python
layer1 = X[0][0]
layer2 = X[0][1]
layer3 = X[0][2]
layer1 = torch.Tensor.cpu(layer1).detach().numpy()[:,-1]
layer2 = torch.Tensor.cpu(layer2).detach().numpy()[:,-1]
layer3 = torch.Tensor.cpu(layer3).detach().numpy()[:,-1]
```

```python
adjacency_matrix = np.load("adjacency_matrix.npy")
```

```python
ego_network = nx.from_numpy_matrix(adjacency_matrix[0])
```

```python
plt.figure(figsize=(14,8))

vmin = layer1.min()
vmax = layer1.max()
cmap = plt.cm.rainbow

nx.draw_networkx(ego_network, with_labels=True, node_color=layer1,
                 cmap=cmap, vmin=vmin, vmax=vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
cbar = plt.colorbar(sm)
```

```python
plt.figure(figsize=(14,8))

vmin = layer2.min()
vmax = layer2.max()
cmap = plt.cm.rainbow

nx.draw_networkx(ego_network, with_labels=True, node_color=layer2,
                 cmap=cmap, vmin=vmin, vmax=vmax)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm.set_array([])
cbar = plt.colorbar(sm)
```
